````python
#----3. Get number of common words in the package name

def char_entropy(text):
    if not text:
        return 0.0
    counts = Counter(text)
    probs = [c / len(text) for c in counts.values()]
    return -sum(p * math.log2(p) for p in probs)

def word_ratio(text):
    words = re.findall(r"[A-Za-z']{2,}", text)
    if not words:
        return 0.0
    known = sum(1 for w in words if zipf_frequency(w.lower(), 'en') > 0)
    return known / len(words)

def non_alpha_ratio(text):
    if not text:
        return 1.0
    alpha = sum(1 for ch in text if ch.isalpha())
    return 1 - (alpha / len(text))

def vowel_ratio(text):
    letters = [c.lower() for c in text if c.isalpha()]
    if not letters:
        return 0.0
    vowels = sum(1 for c in letters if c in 'aeiou')
    return vowels / len(letters)

def looks_like_timestamp_or_number(text):
    if re.match(r'^\d{4}-\d{2}-\d{2}', text):
        return True
    if re.fullmatch(r'\d+(\.\d+)?', text.strip()):
        return True
    return False

def gibberish_score(text):
    e = char_entropy(text)
    wr = word_ratio(text)
    nar = non_alpha_ratio(text)
    vr = vowel_ratio(text)
    ts = looks_like_timestamp_or_number(text)
    score = 0.0
    score += max(0, (e - 4.8)) * 0.4
    score += (1 - wr) * 0.35
    score += nar * 0.15
    score += max(0, (0.35 - vr)) * 0.1
    if ts:
        score = max(score, 0.3)
    return min(1.0, score)

# Convert to string and fill NaN with empty string
data['latest_summary'] = data['latest_summary'].fillna("").astype(str)

# Then apply the gibberish scoring function
data['gibberish_score'] = data['latest_summary'].map(gibberish_score)

# Decide threshold: >0.5 = gibberish
data['likely_gibberish'] = data['gibberish_score'] > 0.5

# Preview
print(data[['latest_summary', 'gibberish_score', 'likely_gibberish']])

````

## Name and description feature engineering
Please look at below code for what i made for name and description raw columns adjustment

````python
import pandas as pd
import numpy as np
import re
from Levenshtein import distance as lev_distance  # Or use RapidFuzz for speed
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def extract_package_features(df, top_legit_names=None, brand_aliases=None, suspicious_keywords=None):
    """
    Extract Package Identity & Naming Features and Summary & Description Features.
    Windows-safe and optimized for medium-sized datasets.
    
    Args:
        df (pd.DataFrame): Must contain 'name', 'latest_summary', 'latest_description'.
        top_legit_names (list): List of top legitimate package names.
        brand_aliases (list): List of brand aliases.
        suspicious_keywords (list): List of suspicious keywords to detect.
        
    Returns:
        pd.DataFrame: Original dataframe with new feature columns added.
    """
    df = df.copy()
    
    # Defaults
    if top_legit_names is None:
        top_legit_names = ['boto3','urllib3','botocore','requests','certifi',
                           'charset-normalizer','idna','typing-extensions','aiobotocore','setuptools']
    if brand_aliases is None:
        brand_aliases = top_legit_names
    if suspicious_keywords is None:
        suspicious_keywords = ['eval','exec','base64','import os','system','subprocess']
    
    # --- Package Identity & Naming Features ---
    df['pkg_name'] = df['name']
    df['n_name_len'] = df['name'].str.len()
    df['has_digit_in_name'] = df['name'].str.contains(r'\d').astype(int)
    df['has_dash_or_underscore'] = df['name'].str.contains(r'[-_]').astype(int)
    
    def name_case(name):
        if name.islower():
            return 'lower'
        elif name[0].isupper():
            return 'camel'
        else:
            return 'mixed'
    df['cat_name_case'] = df['name'].apply(name_case)
    
    # Levenshtein distances
    df['n_lev_dist_to_top1'] = [min([lev_distance(name, t) for t in top_legit_names]) for name in valid_names]
    df['n_lev_dist_to_alias'] = [min([lev_distance(name, t) for t in brand_aliases]) for name in valid_names]
    
    vectorizer = TfidfVectorizer()
    top_vecs = vectorizer.fit_transform(top_legit_names)

# Convert sparse matrix mean to 1D numpy array safely
    centroid = np.asarray(top_vecs.mean(axis=0)).flatten().reshape(1, -1)

# batch-wise transform
    batch_size = 5000
    sim_list = []
    for start in range(0, len(df), batch_size):
        batch_vecs = vectorizer.transform(df['name'].iloc[start:start+batch_size])
        sim_batch = cosine_similarity(batch_vecs.toarray(), centroid)  # convert sparse to dense
        sim_list.append(sim_batch)

    
    df['sim_tfidf_to_legit_centroid'] = np.vstack(sim_list).flatten()
    
    # --- Summary & Description Features ---
    df['n_summary_len'] = df['latest_summary'].fillna('').str.len()
    df['n_desc_len'] = df['latest_description'].fillna('').str.len()
    df['n_desc_lines'] = df['latest_description'].fillna('').str.count('\n') + 1
    df['has_code_block_in_desc'] = df['latest_description'].fillna('').str.contains(r'```').astype(int)
    df['n_urls_in_desc'] = df['latest_description'].fillna('').str.count(r'http[s]?://')
    
    lower_keywords = [k.lower() for k in suspicious_keywords]
    df['has_suspicious_kw'] = df['latest_description'].fillna('').apply(
        lambda x: int(any(k in x.lower() for k in lower_keywords))
    )
    
    def pct_non_ascii(text):
        text = str(text or '')  # convert NaN or float to empty string
        if len(text) == 0:
            return 0.0
        return sum(1 for c in text if ord(c) > 127) / len(text)
    
    df['pct_non_ascii_desc'] = df['latest_description'].apply(pct_non_ascii)

    
    return df
````
